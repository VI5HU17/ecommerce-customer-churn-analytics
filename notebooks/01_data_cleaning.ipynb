```python
import pandas as pd
import os

print("🚀 Starting Data Ingestion Pipeline...")

# 1. Use an active, verified public repository mirror for the Online Retail Dataset
url = "https://raw.githubusercontent.com/shricharan-ks/Retail-datasets/master/Online%20Retail.csv"

# 2. Stream the dataset directly into memory
print("📥 Streaming over 500,000 rows of transactions (This may take a moment)...")
df = pd.read_csv(url)

# 3. Create the absolute target path to save the raw data locally
target_path = os.path.join('..', 'data', 'online_retail.csv')

# 4. Save it as a clean CSV on your computer
df.to_csv(target_path, index=False)

print("\n🎉 SUCCESS!")
print(f"📁 Local File Saved at: {target_path}")
print(f"📊 Matrix Dimensions: {df.shape[0]:,} rows x {df.shape[1]} columns")

print("\n🔍 Baseline Attributes Check:")
print(df.info())
```
